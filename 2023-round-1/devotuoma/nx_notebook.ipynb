# Generators for random graphs

```python
# importing networkx library with the alias nx
import networkx as nx
# importing matplotlib library with the alias plt
import matplotlib.pyplot as plt
```

```python
#creating the graph with 100 nodes and edge probability of 0.3
Random_graph = nx.erdos_renyi_graph(100,0.3)
nx.draw(Random_graph, with_labels=True)
plt.show()
```

# Plotting the degree_centrality distribution using histogram.

```python
def plot_degree_dist(Random_graph):
    degrees = [Random_graph.degree(n) for n in Random_graph.nodes()]
    plt.hist(degrees)
    plt.show()

plot_degree_dist(nx.gnp_random_graph(100, 0.3, directed=True))
```

```python
# Changing the edge probability of the erdos-renyi graph to 0.6 and plotting the degree centrality distribution again.
def plot_degree_dist(Random_graph):
    degrees = [Random_graph.degree(n) for n in Random_graph.nodes()]
    plt.hist(degrees)
    plt.show()

plot_degree_dist(nx.gnp_random_graph(100, 0.6, directed=True))
```

# Inference from the plots
Edge probability act as the bridge between the nodes such that when the probability is high or close to 1 the centrality distribution becomes large. Similarly when the edge probability is far from 1 the centrality distribution becomes smaller. This can be clearly seen as the first histrogram plot with edge probability of 0.3 has half centrality distribution as that of the second plot which has the edge probability of 0.6.
